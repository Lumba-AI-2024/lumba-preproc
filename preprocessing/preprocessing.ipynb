# Cleaning Data

```python
df_pre = df.copy()
```

## duplikat

```python
print("Jumlah duplikasi :" + str(df_pre.duplicated().sum()))
```

```python
df_pre = df_pre.drop_duplicates()
```

```python
print("Jumlah duplikasi :" + str(df_pre.duplicated().sum()))
```

## null

```python
df_pre.isnull().sum()   
```

```python
# drop null df_pre
df_pre = df_pre.dropna()
df_pre
```

# Encoding

```python
#print all unique values in the columns
for col in df_pre.columns:
    print(col, df_pre[col].unique())
```

## ordinal encoding

```python
ordinal_encoding = {'cap_shape' : {'convex': 0, 'bell': 1, 'sunken': 2, 'flat': 3, 'knobbed': 4, 'conical': 5}}

for col, mapping in ordinal_encoding.items():
    df_pre[col] = df_pre[col].map(mapping)
```

## one hot encoding and remove colom with 1 unique value

```python
#drop column with only one unique value
for col in df_pre.columns:
    if len(df_pre[col].unique()) == 1:
        df_pre.drop(col, axis=1, inplace=True)

#one hot encoding for column with more than two unique value
df_pre = pd.get_dummies(df_pre)
df_pre
```

## Label Encode

```python
#label encoding column with only two unique value
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
for col in df_pre.columns:
    if len(df_pre[col].unique()) == 2:
        df_pre[col] = label.fit_transform(df_pre[col])
df_pre
```
